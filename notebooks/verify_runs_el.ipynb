```python
import json
from pathlib import Path
from tqdm import tqdm
from strategic_ttc.verifiers.gpqa_per import GPQAPer

folder = Path("../final_runs/GPQA_el2/")

verifier = GPQAPer()

flag = False

for file in tqdm(list(folder.iterdir())):
    if not file.is_file() or file.suffix not in [".json", ".jsonl"]:
        continue
        
    try:
        with open(file, "r") as f:
            ver_data = [json.loads(line) for line in f if line.strip()]
    except json.JSONDecodeError as e:
        print(f"Skipping corrupt file {file}: {e}")
        continue
    
    new_data = []

    for line in ver_data:
        answers = line.get("answers", [])
        ground_truth = line.get("ground_truths") 
        
        new_explanations = []
        new_correct = []

        for ans in answers:
            result = verifier.verify(
                model_answer=ans,
                ground_truths=ground_truth,
            )
            new_explanations.append(result.explanation)
            new_correct.append(result.correct)

        line["explanations"] = new_explanations
        line["correct"] = new_correct
        new_data.append(line)
        
        if flag == False:
            print(line)
            flag = True
    with open(file, "w") as f:
        for line in new_data:
            f.write(json.dumps(line) + "\n")
    
    flag = False
```
